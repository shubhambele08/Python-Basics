```python
import numpy as np
bulbs = np.random.binomial(1, 0.5, 10)
print("Generated Values:")
print(bulbs)
working = np.sum(bulbs)
print("Number of Working Bulbs:", working)
```

```python
import numpy as np
print("Number of Heads in Each Experiment:")
for i in range(5):
    heads = np.random.binomial(5, 0.5)
    print("Experiment", i + 1, ":", heads)
```

```python
import numpy as np
attempts = np.random.geometric(0.5, 10)
print("Generated Values:")
print(attempts)
print("Maximum Attempts Required:", np.max(attempts))
```

```python
import numpy as np
customers = np.random.poisson(2, 10)
print("Customer Arrivals:")
print(customers)
print("Highest Customers in an Hour:", np.max(customers))
```

```python
import numpy as np
bernoulli = np.random.binomial(1, 0.5, 5)
binomial = np.random.binomial(5, 0.5, 5)
geometric = np.random.geometric(0.5, 5)
poisson = np.random.poisson(2, 5)

print("Bernoulli Distribution:")
print(bernoulli)
print("\nBinomial Distribution:")
print(binomial)
print("\nGeometric Distribution:")
print(geometric)
print("\nPoisson Distribution:")
print(poisson)
```
