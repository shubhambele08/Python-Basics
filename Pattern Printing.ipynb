```python
string=input("Enter a string: ")
n=len(string)
for i in range(n):
    for j in range(i):
        print("*",end="")
    print(string[i],end="")
    for j in range(n-i):
        print("*",end="")
    print()
```

```python
s=input("Enter a string:")
n=len(s)
for i in range(n):
    for j in range(n):
        if i==j or i+j==4:
            print(s[j],end="")
        else:
            print("*",end="")
    print()
```
