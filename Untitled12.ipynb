```python
import numpy as np
```

```python
a=np.arange(1,17).reshape(4,4)
print(a)
```

```python
a[[0,2,3]]
```

```python
a[:3,:3] #All rows except row no 3 ,columns from 0 to 3 excluding 3
```

```python
a[:,0:3] #All rows ,columns from 0 to 3 excluding 3
```

```python
a[0:,0:1] #All rows ,columns no only 0
```

```python
a[0:4,1:2] #All rows except row no 0 ,columns no 1
```

```python
a[3:,0:] #Row from no 3 ,columns from 0 to 3
```
