```python
#Aggregate Functions
a=np.array([10,20,35,40,55])
print(np.sum(a))
print(np.mean(a))
print(np.min(a))
print(np.max(a))
```

```python
a=np.array([10,20,35,40,55])
np.all(a<50)
```

```python
a=np.array([10,20,35,40,55])
b=np.where(a%2!=0,-1,1)
print(b)
```

```python
#Sorting
a=np.array([50,30,60,80,90,100])
b=np.sort(a)
print(b)
```

```python
a=np.array([50,30,60,80,90,100])
a.sort()
b=a.argsort()
print(a)
print(b)
```
