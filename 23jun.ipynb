```python
lst=[1,2,3,4,5,6,7,8,9]
to_search=int(input())
start=0
end=len(lst)-1
mid=(start+end)//2
for i in range(0,len(lst)-1):
  if lst[mid]==to_search:
    print("Found at", {mid} ,"index position")
    break
  elif lst[mid]>to_search:
    end=mid-1
    mid=(start+end)//2
  else:
    start=mid+1
    mid=(start+end)//2
```
